### 5.2. 쯈u칠 elementos de la variable temporal resultan relevantes?

Podr칤amos presentar al modelo toda la informaci칩n incluida en cada marca de tiempo (a침o, mes, d칤a, hora) pero esto podr칤a afectar el entrenamiento pues tendr칤amos muchas variables y no necesariamente todas son relevantes.

La pregunta es entonces, 쯖칩mo sabemos cu치les de estas variables temporales son relevantes al momento de predecir la temperatura?

Una manera de lograr esto es usando la [Transformada de Fourier](https://youtu.be/h4PTucW3Rm0?si=UxDr_kaKP6IAh-jj), una herramienta matem치tica que nos permite conocer los **componentes en frecuencia de una se침al**.

Los componentes en frecuencia son aquellos elementos de la se침al que se repiten con una cierta periodicidad (cada cierto tiempo) y eso es lo que nos interesa encontrar en nuestra se침al.

Entendamos la idea de la Transformada de Fourier con un ejemplo:

> Creemos una se침al cosenoidal, con una frecuencia de 100 Hz (es decir que tiene 100 ciclos en 1 segundo) y veamos su espectro en frecuencia calculado con la Transformada de Fourier.

Comencemos creando la se침al y dibuj치ndola con respecto al tiempo:

```python
import plotly.express as px
import numpy as np

# Crear la se침al
N = 2000                # N칰mero total de muestras (datos): [muestras]
muestras_seg = 4000   # Cantidad de muestras en 1 segundo: [muestras/s]
t = np.linspace(0.0, N/muestras_seg, N ) # Tiempo: desde 0 hasta N/muestras_per_seg = 0.83 s
y = np.cos(100.0 * 2.0*np.pi*t)

# Graficarla con respecto al tiempo
fig = px.line(x=t[0:100], y=y[0:100])
fig.update_layout(xaxis_title='Tiempo (s)')
fig.show()
```

Vemos que la se침al:

- Tiene un periodo de 0.01 segundos
- Lo que equivale a una frecuencia de 1/0.01 Hz = 100 Hz

La idea ahora es llegar a esta misma conclusi칩n pero usando la Transformada de Fourier:

```python
# 1. Calcular la magnitud de la Transformada de Fourier derecha
tf = np.abs(np.fft.rfft(y,N))

# 2. Calcular vector frecuencias (eje horizontal)
frecs = np.fft.rfftfreq(N, d=1./muestras_seg)

# 3. Graficar
fig = px.line(x=frecs, y=tf)
fig.update_layout(xaxis_title=f'Frecuencia (1/s, Hz)')
fig.show()
```

Y vemos que:

- En la gr치fica hay un pico exactamente a una frecuencia de 100 Hz
- Lo que equivale a un periodo de 1/100 = 0.01 s

La gr치fica anterior es lo que se conoce como el **espectro en frecuencia de la se침al**.

Y podemos seguir la misma l칩gica anterior para ver el espectro en frecuencia de la Temperatura.

Comencemos extrayendo esta variable y viendo su comportamiento en el tiempo:

```python
# Importar seaborn
import seaborn as sns
sns.set()

# Extraer la temperatura
temp = df['T (degC)']

# Y graficarla
fig, ax = plt.subplots(figsize=(12,5))
sns.lineplot(temp, ax=ax);
```

Y vemos que efectivamente hay algo de periodicidad en esta Serie de Tiempo.

Pero no resulta f치cil si est치 asociada con el a침o, con el mes, con el d칤a o con la hora. As칤 que podemos usar la Transformada de Fourier para ver esto m치s claramente:

```python
# 0. Par치metros de la serie de tiempo
N = len(temp)      # Cantidad de muestras: [muestras]
muestras_hora = 1            # N칰mero de muestras/horas: [muestras/h]

# 1. Calcular la magnitud de la Transformada de Fourier derecha
tf = np.abs(np.fft.rfft(temp,N))

# 2. Calcular vector frecuencias (eje horizontal)
frecs = np.fft.rfftfreq(N, d=1./muestras_hora)

# 3. Graficar
# Como tenemos 1 muestra por hora pero la serie dura varios a침os
# usaremos una escala logar칤tmica en el eje horizontal
fig = px.line(x=frecs, y=tf, log_x=True)
fig.update_layout(xaxis_title=f'Frecuencia (1/H)')
fig.show()
```

Y en el espectro en frecuencia de nuestra Serie de Tiempo vemos dos componentes predominantes, ubicados en las frecuencias:

1. $119.4648e^{-6}$ [1/H]
2. $0.0416733$ [1/H]

Pero no nos interesa saber la frecuencia sino el equivalente en horas, d칤as, semanas, a침os, etc.

Para hallar esta equivalencia simplemente calculamos el valor inverso (para pasar de frecuencia a tiempo):

```python
frec1 = 119.4648e-6
frec2 = 0.0416733

print(f'La frecuencia {frec1} equivale a {1/frec1} horas')
print(f'La frecuencia {frec2} equivale a {1/frec2} horas')
```

Vemos que la segunda frecuencia equivale pr치cticamente a 24 horas, es decir 1 d칤a.

Por otra parte, la primera frecuencia equivale a 8370.66 horas. Hagamos la conversi칩n a a침os teniendo en cuenta que:

- 1 a침o tiene **en promedio** 365.25 d칤as (para incluir a침os bisiestos)
- 1 d칤a tiene 24 horas

As칤 que, 8370.6665 horas equivalen a:

$8370.6665 \text{ horas} x \frac{1 \text{ d칤a}}{24 \text{ horas}} x \frac{1 \text{ a침o}}{365.25 \text{ d칤as}}$:

```python
resultado = (1/frec1)*(1/24)*(1/365.25)
print(f'La frecuencia {frec1} equivale a {resultado} a침os')
```

춰Es decir pr치cticamente un a침o!

En resumen, este an치lisis nos permite concluir que:

> Las variables d칤a y a침o son relevantes para nuestra Serie de Tiempo.

Y con esto nos podemos enfocar en la manera de codificarlas.

###5.3. 쮺칩mo codificar el a침o y el d칤a?

Requisitos que debe cumplir la codificaci칩n:

1. Debe tener en cuenta el comportamiento c칤clico/peri칩dico de la se침al a predecir
2. Debe haber continuidad en la representaci칩n num칠rica que usemos para el a침o y el d칤a

Si codificamos a침o/d칤a simplemente con los valores originales (2009, 2010, 2011, etc.; 1, 2, 3, ..., 29, 30, etc.) tendremos estos inconvenientes:

1. La representaci칩n num칠rica original no tiene un comportamiento c칤clico (ni para los a침os ni para los d칤as)
2. La representaci칩n num칠rica original contiene saltos (no es continua), especialmente para los d칤as

![](https://drive.google.com/uc?export=view&id=1Got62TEuJKcLHtNKKOBiAft0xfwiJe7K)

Esta representaci칩n dificulta al modelo LSTM encontrar patrones que relacionen las variables temporales con la temperatura.

Una soluci칩n usada com칰nmente en estos casos es **representar las variables temporales con senos y cosenos**, con esto:

1. Se a침ade a la variable temporal un comportamiento c칤clico
2. No habr치 saltos

Por ejemplo, codifiquemos los d칤as usando una funci칩n coseno:

```python
# Codificar cada combinaci칩n AAAA-MM-DD HH:MM:SS a su representaci칩n
# en segundos
tiempo_s = df.index.map(pd.Timestamp.timestamp)
tiempo_s
```

```python
# Por ejemplo, entre dos instantes de tiempo consecutivos
# habr치 exactamente 3.600 segundos = 1 hora!
tiempo_s[2]-tiempo_s[1]
```

```python
# Definir 1 d칤a en segundos como 24 horas x 60 min x 60 s
dia = 24*60*60

# Representar el d칤a senoidal y cosenoidalmente
# La frecuencia de la funci칩n ser치 1/d칤a
dia_sin = np.sin(tiempo_s * (2 * np.pi / dia))

# Graficar 2 d칤as (48 horas)
fig, ax = plt.subplots()
plt.plot(np.array(dia_sin)[:49])
plt.xlabel('Tiempo [horas]')
plt.title('Representaci칩n senoidal del d칤a')

# Dibujar l칤nea horizontal en y = 0.5
ax.axhline(0.25, color='r', linestyle='--', alpha=0.8)

# Dibujar l칤nea vertical en 24 horas
ax.axvline(24, color='k', linestyle='--', alpha=0.8);
```

Vemos que:

1. No hay saltos
2. Hay periodicidad: la hora 0 equivale a la hora 24, la hora 1 a la hora 25, etc.
3. 游녩游녩 Varias horas son codificadas con el mismo valor. Ejemplo: 0h, 10h, 24h son codificadas con 0.25 游녩游녩

Para resolver el problema anterior se puede usar tanto representaci칩n senoidal como cosenoidal:

```python
# Representar la hora con senos y cosenos
dia_sin = np.sin(tiempo_s * (2 * np.pi / dia))
dia_cos = np.cos(tiempo_s * (2 * np.pi / dia))

# Graficar 2 d칤as (48 horas)
fig, ax = plt.subplots()
plt.plot(np.array(dia_sin)[:49])
plt.plot(np.array(dia_cos)[:49])
plt.xlabel('Tiempo [horas]')
plt.title('Representaci칩n senoidal/cosenoidal del d칤a')

# Dibujar l칤nea vertical en 24 horas
ax.axvline(24, color='k', linestyle='--', alpha=0.8);
```

As칤:

> Con la codificaci칩n senoidal y cosenoidal cada instante de tiempo es codificado con un valor 칰nico.

Teniendo esto claro, ahora s칤 podemos codificar las variables temporales (a침o y d칤a) usando senos y cosenos.

Almacenaremos estas codificaciones en nuevas columnas: `dia_sin`, `dia_cos`, `year_sin`, `year_cos`.

Es importante observar que:

> No es necesario combinar d칤as (seno y coseno) y a침os (seno y coseno) en una misma variable. Durante el entrenamiento el modelo LSTM aprender치 a ponderarlas de la manera m치s adecuada.

Este ser칤a el c칩digo requerido:

```python
year = (365.2425)*dia # Valor de un a침o en segundos

df['dia_sin'] = np.sin(tiempo_s * (2 * np.pi / dia))
df['dia_cos'] = np.cos(tiempo_s * (2 * np.pi / dia))
df['year_sin'] = np.sin(tiempo_s * (2 * np.pi / year))
df['year_cos'] = np.cos(tiempo_s * (2 * np.pi / year))
df
```

Tengamos en cuenta que **NO ser치 necesario escalar las 4 variables que acabamos de crear, pues ya est치n en el rango de -1 a 1**.

Para simplificar el *DataFrame* podemos quitar el 칤ndice con las marcas temporales, pues ya est치n codificadas (esto es opcional):

### 7.1. Partici칩n del set en entrenamiento, validaci칩n y prueba

Esta fase es id칠ntica a la implementada en el modelo anterior (multivariado - unistep).

A diferencia de otros modelos de Machine Learning, en el caso de series de tiempo **se debe garantizar que se generan las particiones sin mezclar aleatoriamente los datos**.

Recordemos adem치s que:
- El set de entrenamiento (*train*) se usar치 para encontrar los par치metros del modelo
- El set de validaci칩n (*val*) para verificar que no haya *under/over-fitting* del modelo y para ajustar sus hiperpar치metros
- El set de prueba (*test*) para poner a prueba el mejor modelo encontrado durante el entrenamiento/validaci칩n

En este caso usaremos la misma funci칩n implementada para los modelos univariados con la diferencia de que en lugar de introducir una *serie* de Pandas, introduciremos el *DataFrame* completo.

Por tanto, la funci칩n retornar치 tres *dataframes* (train, val y test):

```python
# Funci칩n para generar las particiones preservando las caracter칤sticas
# de la serie de tiempo

def train_val_test_split(dataframe, tr_size=0.8, vl_size=0.1, ts_size=0.1 ):
    # Definir n칰mero de datos en cada subserie
    N = dataframe.shape[0]
    Ntrain = int(tr_size*N)  # N칰mero de datos de entrenamiento
    Nval = int(vl_size*N)    # N칰mero de datos de validaci칩n
    Ntst = N - Ntrain - Nval # N칰mero de datos de prueba

    # Realizar partici칩n
    train = dataframe[0:Ntrain]
    val = dataframe[Ntrain:Ntrain+Nval]
    test = dataframe[Ntrain+Nval:]

    return train, val, test

# Prueba de la funci칩n
tr, vl, ts = train_val_test_split(df)

print(f'Tama침o set de entrenamiento: {tr.shape}')
print(f'Tama침o set de validaci칩n: {vl.shape}')
print(f'Tama침o set de prueba: {ts.shape}')
```

Observemos que en lugar de una columna cada subset tiene ahora 17 (correspondientes a las covariables).

###7.2 Generaci칩n del dataset supervisado (entrada y salida del modelo)

En este caso lo 칰nico que cambia con respecto al tutorial anterior es que ahora tendremos 17 *features* de entrada al modelo: las 13 originales + 4 variables temporales.

Debemos ajustar nuestro set de datos de acuerdo a lo especificado en la [documentaci칩n de las Redes LSTM en TensorFlow/Keras](https://www.tensorflow.org/api_docs/python/tf/keras/layers/LSTM):

![](https://drive.google.com/uc?export=view&id=1GnrHTdwbIRqjxm0xpkwWetCrTs_MMEqq)

As칤 que en este caso cada dato de entrenamiento ser치:

- Un arreglo de 24 (horas) x 17 (features) correspondiente a la entrada
- Un arreglo de 4 (horas) x 1 (feature):

![](https://drive.google.com/uc?export=view&id=1GoZ9hpf9q9Q-tFUi6kYtBMcWXjcJyNGv)

En este caso, `INPUT_LENGTH` (24 horas) y `OUTPUT_LENGTH`(4 horas) se convertir치n en hiperpar치metros que m치s adelante se deber칤an afinar para mejorar el desempe침o del modelo.

Y usaremos la misma funci칩n usada en los tutoriales anteriores:

```python
def crear_dataset_supervisado(array, input_length, output_length):
    '''Permite crear un dataset con las entradas (X) y salidas (Y)
    requeridas por la Red LSTM.

    Par치metros:
    - array: arreglo numpy de tama침o N x features (N: cantidad de datos,
      f: cantidad de features)
    - input_length: instantes de tiempo consecutivos de la(s) serie(s) de tiempo
      usados para alimentar el modelo
    - output_length: instantes de tiempo a pronosticar (salida del modelo)
    '''

    # Inicializaci칩n
    X, Y = [], []    # Listados que contendr치n los datos de entrada y salida del modelo
    shape = array.shape
    if len(shape)==1: # Si tenemos s칩lo una serie (univariado)
        fils, cols = array.shape[0], 1
        array = array.reshape(fils,cols)
    else: # Multivariado <-- <--- 춰esta parte de la funci칩n se ejecuta en este caso!
        fils, cols = array.shape

    # Generar los arreglos
    for i in range(fils-input_length-output_length):
        X.append(array[i:i+input_length,0:cols])
        Y.append(array[i+input_length:i+input_length+output_length,-1].reshape(output_length,1))

    # Convertir listas a arreglos de NumPy
    X = np.array(X)
    Y = np.array(Y)

    return X, Y
```

```python
# Crear los datasets de entrenamiento, prueba y validaci칩n y verificar sus tama침os
INPUT_LENGTH = 24    # Hiperpar치metro
OUTPUT_LENGTH = 4    # Modelo multi-step

x_tr, y_tr = crear_dataset_supervisado(tr.values, INPUT_LENGTH, OUTPUT_LENGTH)
x_vl, y_vl = crear_dataset_supervisado(vl.values, INPUT_LENGTH, OUTPUT_LENGTH)
x_ts, y_ts = crear_dataset_supervisado(ts.values, INPUT_LENGTH, OUTPUT_LENGTH)

# Imprimir informaci칩n en pantalla
print('Tama침os entrada (BATCHES x INPUT_LENGTH x FEATURES) y de salida (BATCHES x OUTPUT_LENGTH x FEATURES)')
print(f'Set de entrenamiento - x_tr: {x_tr.shape}, y_tr: {y_tr.shape}')
print(f'Set de validaci칩n - x_vl: {x_vl.shape}, y_vl: {y_vl.shape}')
print(f'Set de prueba - x_ts: {x_ts.shape}, y_ts: {y_ts.shape}')
```

```python
from sklearn.preprocessing import MinMaxScaler

def escalar_dataset(data_input, col_ref):
    '''Escala el dataset en el rango de -1 a 1.

    Entradas:
    - data_input: diccionario con los dataset de entrada y salida del modelo
    (data_input = {'x_tr':x_tr, 'y_tr':y_tr, 'x_vl':x_vl, 'y_vl':y_vl,
                    'y_ts':y_ts})
    - col_ref: par치metro adicional para especificar la columna que contiene
      la variable a predecir


    Retorna:
    - data_scaled: diccionario con los datasets de entrada y salida escalados
      (tiene la misma estructura del diccionario de entrada)
    - scaler: el escalador usado (requerido para las predicciones)
    '''

    # *** Implementaci칩n adicional: determinar el 칤ndice de la columna
    # que contiene la variable a predecir
    col_ref = df.columns.get_loc(col_ref)

    # N칰mero de instantes de tiempo de entrada y de covariables
    NSAMPLES = data_input['x_tr'].shape[1]
    NFEATS = data_input['x_tr'].shape[2]

    # Generar listado con "scalers" (1 por cada covariable de entrada)
    scalers = [MinMaxScaler(feature_range=(-1,1)) for i in range(NFEATS)]

    # Arreglos que contendr치n los datasets escalados
    x_tr_s = np.zeros(data_input['x_tr'].shape)
    x_vl_s = np.zeros(data_input['x_vl'].shape)
    x_ts_s = np.zeros(data_input['x_ts'].shape)
    y_tr_s = np.zeros(data_input['y_tr'].shape)
    y_vl_s = np.zeros(data_input['y_vl'].shape)
    y_ts_s = np.zeros(data_input['y_ts'].shape)

    # Escalamiento: se usar치n los min/max del set de entrenamiento para
    # escalar la totalidad de los datasets

    # *************************************************
    # Escalamiento Xs: porci칩n de c칩digo modificada para no escalar las
    # columnas con las variables temporales (columnas 11->14)
    # *************************************************
    FEAT_IDS = list(range(11)) + list(range(15,17)) # Columnas 0->10 + 15-16
    for i in FEAT_IDS:
        x_tr_s[:,:,i] = scalers[i].fit_transform(x_tr[:,:,i])
        x_vl_s[:,:,i] = scalers[i].transform(x_vl[:,:,i])
        x_ts_s[:,:,i] = scalers[i].transform(x_ts[:,:,i])

    # *************************************************
    # Y dejar las columnas con variables temporales (13->16) intactas
    # *************************************************
    FEATS_TIME = list(range(11,15)) # Columnas 11-14
    x_tr_s[:,:,FEATS_TIME] = x_tr[:,:,FEATS_TIME]
    x_vl_s[:,:,FEATS_TIME] = x_vl[:,:,FEATS_TIME]
    x_ts_s[:,:,FEATS_TIME] = x_ts[:,:,FEATS_TIME]


    # Escalamiento Ys (teniendo en cuenta "col_ind")
    y_tr_s[:,:,0] = scalers[col_ref].fit_transform(y_tr[:,:,0])
    y_vl_s[:,:,0] = scalers[col_ref].transform(y_vl[:,:,0])
    y_ts_s[:,:,0] = scalers[col_ref].transform(y_ts[:,:,0])

    # Conformar diccionario de salida
    data_scaled = {
        'x_tr_s': x_tr_s, 'y_tr_s': y_tr_s,
        'x_vl_s': x_vl_s, 'y_vl_s': y_vl_s,
        'x_ts_s': x_ts_s, 'y_ts_s': y_ts_s,
    }

    return data_scaled, scalers[col_ref]
```

```python
# Escalamiento del dataset con la funci칩n anterior

# Crear diccionario de entrada
data_in = {
    'x_tr': x_tr, 'y_tr': y_tr,
    'x_vl': x_vl, 'y_vl': y_vl,
    'x_ts': x_ts, 'y_ts': y_ts,
}

# Y escalar (especificando la columna con la variable a predecir)
data_s, scaler = escalar_dataset(data_in, col_ref = 'T (degC)' )

# Extraer subsets escalados
x_tr_s, y_tr_s = data_s['x_tr_s'], data_s['y_tr_s']
x_vl_s, y_vl_s = data_s['x_vl_s'], data_s['y_vl_s']
x_ts_s, y_ts_s = data_s['x_ts_s'], data_s['y_ts_s']
```

Usaremos la misma funci칩n `predecir` usada en los videos anteriores:

```python
def predecir(x, model, scaler):
    '''Genera la predicci칩n de OUTPUT_LENGTH instantes
    de tiempo a futuro con el modelo entrenado.

    Entrada:
    - x: batch (o batches) de datos para ingresar al modelo
      (tama침o: BATCHES X INPUT_LENGTH X FEATURES)
    - model: Red LSTM entrenada
    - scaler: escalador (requerido para llevar la predicci칩n a la escala original)

    Salida:
    - y_pred: la predicci칩n en la escala original (tama침o: BATCHES X OUTPUT_LENGTH X FEATURES)
    '''

    # Calcular predicci칩n escalada en el rango de -1 a 1
    y_pred_s = model.predict(x,verbose=0)

    # Llevar la predicci칩n a la escala original
    y_pred = scaler.inverse_transform(y_pred_s)

    return y_pred.flatten()
```
